```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
from scipy.stats import linregress
%matplotlib inline

tesla = pd.read_excel(r'C:\Users\timod\Desktop\TESLA_3M.xlsx')
tsla = pd.read_excel(r'C:\Users\timod\Desktop\TSLA_3M.xlsx')
tslaHD = pd.read_excel(r'C:\Users\timod\Desktop\TSLA_HD_3M.xlsx')

print(tesla.shape)
print(tsla.shape)
print(tslaHD.shape)
```
